```python
#Write your python code here
def sumArray(arr, size):
    total = 0
    for i in range(size):
        total += arr[i]
    return total
size = 5
numbers = []
print("Enter",size,"numbers: ")
for i in  range(size):
   number= int(input())
   numbers.append(number)
sum = sumArray(numbers,size)
print("Sum of the elements: ", sum)
```
