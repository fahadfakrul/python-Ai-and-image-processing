```python
# Import the modules from skimage
from skimage import color, data, io
import matplotlib.pyplot as plt

img = io.imread("/content/fruits-2.jpg")

# Convert the image to grayscale
gray_scaled_img = color.rgb2gray(img)

# Show the original image
plt.imshow(img)
plt.title('Original RGB image')
plt.axis('off')  # Hide axes
plt.show()

# Display the grayscale image
plt.imshow(gray_scaled_img, cmap='gray')
plt.title('Grayscale image')
plt.axis('off')  # Hide axes
plt.show()
```

```python
prank_img= io.imread("/content/sevilleup(2).jpg")
import numpy as np
plt.imshow(prank_img)
plt.axis('off')
plt.show()
seville_vertical_flip = np.flipud(prank_img)
plt.imshow(seville_vertical_flip)
plt.axis('off')
plt.show()
```

```python
# Flip the previous image horizontally
seville_horizontal_flip = np.fliplr(seville_vertical_flip)
plt.imshow(seville_horizontal_flip)
plt.axis('off')
plt.show()
```
